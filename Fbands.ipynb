```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f_band = np.genfromtxt('Fstat_noise_fband.txt', usecols = (0,6), comments = '%')
f_band_shift = np.genfromtxt('Fstat_noise_fband_shift.txt', usecols = (0,6), comments = '%')

f1_band = np.genfromtxt('Fstat_noise_f1band.txt', usecols = (3,6), comments = '%')
f2_band = np.genfromtxt('Fstat_noise_f2band.txt', usecols = (4,6), comments = '%')
```

```python
pathTimeStampsH1_O2_1 = '/home/m206265/lalapps_work/data/O2_1_H1_tslist.txt'
TimeStampsH1_O2_1 = np.genfromtxt(pathTimeStampsH1_O2_1)

pathTimeStampsL1_O2_1 = '/home/m206265/lalapps_work/data/O2_1_L1_tslist.txt'
TimeStampsL1_O2_1 = np.genfromtxt(pathTimeStampsL1_O2_1)

TimeStampsL1H1_O2_1 = np.concatenate((TimeStampsH1_O2_1, TimeStampsL1_O2_1))

Tstart = min(TimeStampsL1H1_O2_1)
Tend = max(TimeStampsL1H1_O2_1)
T_obs = (Tend - Tstart)+1800
Freq=60.001
dF = 1/T_obs
print(dF)
```

```python
fig, ax1 = plt.subplots(1,1)
ax1.plot((f_band[:,0]-Freq)/dF, f_band[:,1], '.')
ax1.plot((f_band_shift[:,0]-Freq)/dF, f_band_shift[:,1], '.')
ax1.set_xlabel('Frequency template grid')
ax1.set_ylabel('2F')
ax1.grid()
#plt.show()
plt.savefig('crabfband.pdf')
```
